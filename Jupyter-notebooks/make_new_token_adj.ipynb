```python
def make_adj_from_token_seq(token_seq_list, num_tokens, window_size = 2):

    token_adj = np.zeros((num_tokens, num_tokens))

    for tok_seq in token_seq_list:
        # get 100 random samples from tokens
        # tok_indices = np.random.randint(low = 0, high = len(tok_seq) - 1, size = num_token_samples)
        tok_range = range(window_size, len(tok_seq) - window_size, window_size + 1) #np.linspace(window_size, 3)
        # tok_indices = np.random.randint(low = 0, high = len(tok_seq) - 1, size = len(tok_seq) - 1)

        for i in tok_range:
            token_adj[tok_seq[i], tok_seq[i]] += 1
            # direct neighbors
            token_adj[tok_seq[i], tok_seq[i+1]] += 1
            token_adj[tok_seq[i], tok_seq[i-1]] += 1

            token_adj[tok_seq[i], tok_seq[i+2]] += 0.5
            token_adj[tok_seq[i], tok_seq[i-2]] += 0.5
    return token_adj

            
```

```python
adj_out = make_adj_from_token_seq(token_seq_list=new_tokens, num_tokens = tok_emb.shape[0])
```
